## Teoría del Confort Térmico en Ganado Vacuno

El confort térmico en el ganado vacuno se refiere a un estado en el que los animales no experimentan estrés por calor ni por frío, permitiéndoles mantener su temperatura corporal central dentro de un rango fisiológico normal con un gasto mínimo de energía. Este estado es crucial para la salud, el bienestar y la productividad de los animales (producción de leche, carne, reproducción).

Los bovinos son homeotermos, lo que significa que mantienen una temperatura corporal constante independientemente de la temperatura ambiente. Sin embargo, su capacidad para hacerlo está limitada por el rango de temperaturas al que están expuestos y por factores intrínsecos (edad, raza, nivel de producción) y extrínsecos (humedad, radiación solar, velocidad del viento).

Cuando las condiciones ambientales se desvían de su **zona de termoneutralidad** (rango de temperaturas donde el animal no necesita gastar energía adicional para mantener su temperatura corporal), los animales experimentan **estrés térmico**. Este estrés puede ser por calor (cuando la temperatura ambiente es demasiado alta) o por frío (cuando es demasiado baja), siendo el estrés por calor el más común en muchas regiones tropicales y subtropicales.

### Estrés por Calor

El estrés por calor ocurre cuando la carga de calor en el animal (proveniente del ambiente y de su propio metabolismo) excede su capacidad para disipar ese calor. Para compensar, los animales activan mecanismos fisiológicos y conductuales:

*   **Fisiológicos:** Aumento de la frecuencia respiratoria (jadeo), aumento de la sudoración (limitada en bovinos), vasodilatación periférica para aumentar el flujo sanguíneo a la piel y disipar calor.
*   **Conductuales:** Búsqueda de sombra, reducción de la actividad, aumento del consumo de agua, disminución del consumo de alimento.

Consecuencias del estrés por calor:

*   Disminución del consumo de alimento y, consecuentemente, de la producción de leche y carne.
*   Reducción de la eficiencia reproductiva (disminución de la tasa de concepción, aumento de la mortalidad embrionaria).
*   Debilitamiento del sistema inmunitario, aumentando la susceptibilidad a enfermedades.
*   Acidosis, cojeras y otros problemas de salud.
*   En casos extremos, puede llevar a la muerte.

### Índices de Confort Térmico

Debido a que la temperatura ambiente por sí sola no es suficiente para evaluar el estrés térmico, se han desarrollado índices que combinan múltiples variables meteorológicas. El **Índice de Temperatura y Humedad (ITH)** es el más ampliamente utilizado y combina la temperatura del aire con la humedad relativa, ya que la humedad reduce la capacidad del animal para disipar calor por evaporación.

Los valores de ITH se correlacionan con diferentes niveles de estrés térmico, permitiendo a los productores identificar situaciones de riesgo y aplicar medidas de mitigación oportunas, como proporcionar sombra, acceso a agua fresca, ventilación, o sistemas de enfriamiento.

# *Métodos y cálculos*

* Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72(T_{bs} + T_{bh}) + 40.6$$

Donde:


$T_{bs}$ Temperatura bulbo seco ($^oC$)


$T_{bh}$ Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

Sí $ITH ≤ 70$ ambiente no estresante

Sí $ITH > 70 \space And \space ITH ≤ 78$ situación de alerta por estrés térmico

Sí $ITH > 78 \space And \space ITH ≤ 83$ peligro por estrés térmico

Sí $ITH > 83$ situación crítica

````python
# ENTRADA: Pedir tbs (Temperatura de bulbo secoo) y tbh (Temperatura de bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados Centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados Centigrados: "))

# PROCESO: Calculo de ith (índice de Temperatura y Humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: Imprimir resultado de la evaluación del ith y el confort térmico
if ith <= 70:
  print("El índice de temperatura y humedad es: ", ith)
  print("Ambiente no estresante")

elif ith > 70 and ith <= 78:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación de alerta por estrés térmico")

elif ith > 78 and ith <= 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("peligro por estrés térmico")

elif ith > 83:
  print("El índice de temperatura y humedad es: ", ith)
  print("Situación crítica")
````
