### Un ejemplo aplicado a la Ingeniería

En lugar de un simple "Hola Mundo", veamos algo que podrías usar en clase. Supongamos que necesitamos calcular el **volumen de agua** necesario para llenar un tanque cilíndrico de almacenamiento.

Antes de mostrarte el código, analicemos la lógica:

1. Necesitamos el **radio** y la **altura** (variables).
2. Usamos la fórmula matemática: $V = \pi \cdot r^2 \cdot h$.
3. Mostramos el resultado.

Aquí tienes cómo se vería un script básico y bien estructurado:

```python
import math

def calcular_volumen_tanque(radio_m, altura_m):
    """
    Calcula el volumen de un tanque cilíndrico en metros cúbicos.
    """
    # Usamos la constante pi de la librería math
    volumen = math.pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada con nombres descriptivos
radio_tanque = 2.5  # metros
altura_tanque = 5.0 # metros

# Llamada a la función
resultado_v = calcular_volumen_tanque(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {resultado_v:.2f} m3")
```

---

### ¿Por qué este código es "bueno"?

* **Modularización:** He creado una `def calcular_volumen_tanque`. Esto permite que, si mañana tienes 100 tanques, solo llames a la función en lugar de repetir la fórmula.
* **Claridad:** Nota que no usé `r` o `h`, sino `radio_m` y `altura_m`. En ingeniería, saber la **unidad de medida** en el nombre de la variable evita desastres.
