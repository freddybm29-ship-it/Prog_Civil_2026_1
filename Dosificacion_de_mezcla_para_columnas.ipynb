# **Dosificación de Mezcla para Columnas**

**Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v= \pi * r^2 * h$$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$Sacos=v * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

````python
#[ENTRADA]: Solicitar la altura de la columna en metros
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: imprime el resultado
print("Volumen ≈", round(v, 3), "m3")
print("Sacos necesarios ≈", round(sacos, 2), "sacos")
````

```markdown
El siguiente código calcula la cantidad de sacos de cemento necesarios para una columna, basado en su altura y diámetro:

*   `h = float(input("Ingrese la altura de la columna en metros: "))`
    *   Esta línea solicita al usuario que ingrese la altura de la columna en metros. El valor ingresado se convierte a un número de punto flotante (`float`) y se almacena en la variable `h`.

*   `d = float(input("Ingrese el diámetro de la columna en metros: "))`
    *   De manera similar, esta línea pide al usuario el diámetro de la columna en metros. El valor se convierte a un número de punto flotante y se guarda en la variable `d`.

*   `r = d/2`
    *   Aquí se calcula el radio (`r`) de la columna, dividiendo el diámetro (`d`) entre 2.

*   `v = 3.1416 * r**2 * h`
    *   Esta línea calcula el volumen (`v`) de la columna cilíndrica utilizando la fórmula \( V = \pi \cdot r^2 \cdot h \). Se usa `3.1416` como una aproximación de \( \pi \) (pi), `r**2` para elevar el radio al cuadrado, y se multiplica por la altura `h`.

*   `sacos = v * 9.5`
    *   Finalmente, se calcula la cantidad total de sacos de cemento (`sacos`) multiplicando el volumen (`v`) de la columna por la constante de consumo de cemento (9.5 sacos por metro cúbico).

*   `print("Volumen ≈", round(v, 3), "m3")`
    *   Esta línea imprime el volumen calculado. La función `round(v, 3)` redondea el volumen a tres decimales antes de mostrarlo.

*   `print("Sacos necesarios ≈", round(sacos, 2), "sacos")`
    *   Esta última línea imprime la cantidad de sacos de cemento necesarios. La función `round(sacos, 2)` redondea la cantidad de sacos a dos decimales para una presentación más legible.
```
